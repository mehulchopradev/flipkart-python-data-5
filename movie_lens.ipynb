```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movielens/movies.dat', sep='::', header=None, names=['movie_id', 'title', 'genres'])
movies
```

```python
users = pd.read_csv('movielens/users.dat', sep='::', header=None, names=['user_id', 'gender', 'age', 'occup', 'zip'])
users
```

```python
ratings = pd.read_csv('movielens/ratings.dat', sep='::', header=None, names=['user_id', 'movie_id', 'ratings', 'timestamp'])
ratings
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
d1 = data[['title', 'gender', 'ratings']]
d1
```

```python
d1['title'].value_counts(ascending=True)
```

```python
s1 = d1['title']
s2 = s1.value_counts()
s3 = s2[s2 > 500]
popular_films = s3.index
popular_films
```

```python
d1
```

```python
d2 = d1.pivot_table('ratings', aggfunc='mean', index='title', columns=['gender'])
d2
```

```python
d3 = d2.loc[popular_films]
d3
```

```python
d3.sort_values(by='F', ascending=False).head(10)
```

```python
d3.sort_values(by='M', ascending=False).head(10)
```

```python
d3['diff'] = d3['F'] - d3['M']
d3['diff'] = d3['diff'].abs()
```

```python
d4 = d3.sort_values(by='diff', ascending=False).head(10)
d4
```

```python
d5 = d4.drop(columns=['F', 'M'])
d5
```

```python
d5.plot(kind='bar')
```
